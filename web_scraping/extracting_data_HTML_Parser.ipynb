```python
import requests
import json
from bs4 import BeautifulSoup
```

```python
def scrape_and_store(url , storage_file):
    try:
        
        response = requests.get(url)
#         print(response.text)
        
        if response.status_code == 200:
            #parse 
            soup = BeautifulSoup(response.text, "html.parser")
            print(soup.text)
            
            data_to_store = {
                "h1":soup.title.string.strip(),
                "url":url,
                "content":soup.get_text().strip()
            }
            
            
            
            with open(storage_file , 'a') as f:
                json.dump(data_to_store, f)
                f.write('\n')
                
            print(f"data from {url} stored successful.")
        
        else:
            print(f"Failed to retrieve data from {url}")
    
    except Exception as e:
        print(f"An error occured: {str(e)}")
        
            
        
#example usage 
start_url = "http://quotes.toscrape.com/js"
storage_file = 'scrapped_data.json'


            
```

```python
scrape_and_store(start_url , storage_file)
```
